```python
import pandas as pd
import plotly.express as px
```

```python
unemployment_path = "../dataset/Unemployment/"

unemployment_df = pd.read_csv(unemployment_path + "Unemployment_Rate_M_Total.txt").dropna()
```

```python
unemployment_df
```

```python
unemployment_df_other = pd.read_csv(unemployment_path + "Unemployment_Rate_upto_11_2020.csv").dropna()
```

```python
region_set=set()
```

```python
for i in unemployment_df_other['Region']:
    region_set.add(i)
```

```python
region_set
```

```python
unemployment_df_other.columns
```

```python
latitudes=[]
longitudes=[]
for state in region_set:
    latitude=unemployment_df_other.loc[unemployment_df_other['Region'] == state,'latitude' ].iloc[0]
    longitude=unemployment_df_other.loc[unemployment_df_other['Region'] == state,'longitude' ].iloc[0]
    latitudes.append(latitude)
    longitudes.append(longitude)
```

```python
latitude_region_dict={}
longitude_region_dict={}
i=0
for state in region_set:
    latitude_region_dict[state]=latitudes[i]
    longitude_region_dict[state]=longitudes[i]
    i+=1
```

```python
unemployment_df1=unemployment_df[unemployment_df['Region']!='India']
```

```python
latitude_column=[]
longitude_column=[]

for state in unemployment_df1['Region']:
    latitude=latitude_region_dict[state]
    latitude_column.append(latitude)
    longitude=longitude_region_dict[state]
    longitude_column.append(longitude)
```

```python
unemployment_df1['latitude']=latitude_column
unemployment_df1['longitude']=longitude_column
```

```python
unemployment_df1
```

```python
def get_animation(unemployment_df_other):
    unemployment_df_other1=unemployment_df_other.copy()
    unemployment_df_other1[' Date'] = pd.to_datetime(unemployment_df_other1[' Date'],dayfirst=True)

    unemployment_df_other1['Year'] =  unemployment_df_other1[' Date'].dt.year

    fig = px.scatter_geo(unemployment_df_other1,'longitude', 'latitude', color="Region",
                     hover_name="Region", size=" Estimated Unemployment Rate (%)",
                     animation_frame="Year",scope='asia',template='plotly',title='Unemployment Rate in India from 2016 to 2022')
    
    frames = []
    for s, fr in enumerate(fig.frames):
        # set main traces to appropriate traces within plotly frame
        fig.update(data=fr.data)
        # move slider to correct place
        fig.layout.sliders[0].update(active=s)
        # generate image of current state
        frames.append(PIL.Image.fromopen(io.BytesIO(fig.to_image(format="png"))))

    # create animated GIF
    frames[0].save(
            "test.gif",
            save_all=True,
            append_images=frames[1:],
            optimize=True,
            duration=500,
            loop=0,
        )
    fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 2000

    fig.update_geos(lataxis_range=[5,35], lonaxis_range=[65, 100],oceancolor="#6dd5ed", showocean=True)
    fig.show()
```

```python
get_animation(unemployment_df1)
```
